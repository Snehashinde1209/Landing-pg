```
top_10 = df['make'].value_counts().nlargest(10)
total = df['make'].value_counts().sum()
others = total - top_10.sum()
top_10['others'] = others

plt.title('Car sales by brand', fontweight = 'bold')
plt.pie(top_10, labels = top_10.index, autopct = '%1.1f%%', wedgeprops = dict(edgecolor = 'black'));
```

```
color_analysis = df['color'].value_counts().reset_index(name='count')
color_analysis.columns = ['color', 'count']

plt.title('Color preference', fontweight='bold')
plt.bar(color_analysis['count'],
        color_analysis['color'], edgecolor = 'blue')
plt.xlabel('color')
plt.ylabel('Cars saled')
plt.xticks(rotation = 45);
```

```

df['profit'] = df['sellingprice'] - df['mmr']
df['profit'].describe()
```

```

plt.figure(figsize = (20,8))
plt.title('Profit of every sale, by color', fontweight = 'bold')
sns.violinplot(x = 'color', y = 'profit', data = df, hue = 'color')
plt.xticks(rotation = 45)
plt.xlabel('color')
plt.ylabel('Profit (in $)');
```
